## 8Bits

```python
from transformers import BitsAndBytesConfig
from transformers import AutoModelForCausalLM, AutoTokenizer


config = BitsAndBytesConfig(load_in_4bit=True)
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-Math-7B", quantization_config=config)
```

```python
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-Math-7B")

def generate(prompt):
    tokenized_text = tokenizer(prompt, return_tensors='pt').to('cuda')
    output = model.generate(**tokenized_text)
    result = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
    return result
```

```python
generate('Quanto é e^1 + 1?')
```
