```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/2024-09-11T13-20_export.csv')
data.Color = data['Color'].map({'red': 'A', 'blue': 'B'})
sns.scatterplot(x=data.X,y=data.Y, hue=data['Color'])
plt.grid()
```
