```python
# data
# w
# multiply X by w
# apply threshold

# Score
    # create a random vector with "w" values
    # multiply each xi for wi
    # determine threshold
    # show output

def simple_model(X, Y, w):
    score = []
    sum_score = []

    for i in X:
        score.append(((((i*w)))))

    for i in (score):
        sum_score.append(sum(i))

    for i in sum_score:
        if np.sign(i) == 1:
            print(f"Approved, Score: {i} threshold {np.sign(i)}")

        else:
            print(f"Rejected, Score: {i} threshold {np.sign(i)}")

w = np.array([0.1, 0.5, 0.8, 0.11, 0.12])
simple_model(X, Y, w)
```
