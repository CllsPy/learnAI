## [The pipeline function](https://huggingface.co/docs/transformers/main_classes/pipelines)

```python
from transformers import pipeline
```

```python
%%capture
classifier = pipeline("sentiment-analysis")
```

```python
classifier ([

               "I hate you!",
               "I love you!",
               "You make me wanna die!"
])
```

```python
%%capture
## zero-shot-classification
classifier2 = pipeline("zero-shot-classification")
```

```python
classifier2 (

               "I love machine learning soo much!",
               candidate_labels=["politics", "education", "business"]

)
```

```python
## generation
%%capture
generator = pipeline("text-generation")
```

```python
generator("Today is..")
```

```python
%%capture
## gpt-2
generator2 = pipeline("text-generation", model="distilgpt2")
```

```python
generator2(
    "In this class will teach you how to",
    max_length=30,
    num_return_sequences=1
)
```

```python
%%capture
## mask
unmasker = pipeline("fill-mask")
```

```python
unmasker(
    "This course will teach you <mask>",
    top_k = 2

)
```

```python
%%capture
## ner
ner = pipeline("ner", grouped_entities=True)
```

```python
ner ("My name is Carlos I live in Brazil and I Work on HuggingFace")
```

```python
%%capture
# Q&A
question_answering = pipeline("question-answering")
```

```python
question_answering(
    question="Where do I live?",
    context="I'm live in Brazil"

)
```
