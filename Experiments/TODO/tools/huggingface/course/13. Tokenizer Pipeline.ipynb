```python
from transformers import AutoTokenizer
```

### Bert


```python
%%capture
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
inputs = tokenizer("I love huggingface!")
print(inputs["input_ids"])
```

```python
tokens = tokenizer.tokenize("I love huggingface!")
print(tokens)
```

```python
inputs_id = tokenizer.convert_tokens_to_ids(tokens)
print(inputs_id)
```

```python
final_inputs = tokenizer.prepare_for_model(inputs_id)
print(final_inputs["input_ids"])
```

```python
print(tokenizer.decode(final_inputs["input_ids"]))
```

```python
one_step = tokenizer("I love huggingface!")
print(one_step)
```
